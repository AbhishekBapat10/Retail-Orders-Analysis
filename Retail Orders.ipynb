```python
query = "SELECT * FROM orders LIMIT 5"
result = pd.read_sql(query, conn)
print(result)
```

```python
# Q1) Top 10 highest revenue generating products
q1 = '''
SELECT product_id, SUM(sale_price) AS sales
FROM orders
GROUP BY product_id
ORDER BY sales DESC
LIMIT 10
'''
result = pd.read_sql(q1, conn)
print(result)
```

```python
#Q2) Find top highest selling products in each region
q2='''WITH cte AS (
    SELECT region, product_id, SUM(sale_price) AS sales
    FROM orders
    GROUP BY region, product_id
)
SELECT region, product_id, sales
FROM (
    SELECT region, product_id, sales,
           ROW_NUMBER() OVER(PARTITION BY region ORDER BY sales DESC) AS rn
    FROM cte
) AS A
WHERE rn <= 5;'''

result = pd.read_sql(q2, conn)
print(result)
```

```python
#Q3)Find month over month growth comparison for 2022 and 2023 sales eg : Jan 2022 vs Jan 2023
q3='''WITH cte AS (
    SELECT strftime('%Y', order_date) AS order_year,
           strftime('%m', order_date) AS order_month,
           SUM(sale_price) AS sales
    FROM orders
    GROUP BY strftime('%Y', order_date), strftime('%m', order_date)
)
SELECT order_month,
       SUM(CASE WHEN order_year = '2022' THEN sales ELSE 0 END) AS sales_2022,
       SUM(CASE WHEN order_year = '2023' THEN sales ELSE 0 END) AS sales_2023
FROM cte
GROUP BY order_month
ORDER BY order_month;
'''
result = pd.read_sql(q3, conn)
print(result)
```

```python
q4='''WITH cte AS (
    SELECT strftime('%Y', order_date) AS order_year,
           strftime('%m', order_date) AS order_month,
           SUM(sale_price) AS sales
    FROM orders
    GROUP BY strftime('%Y', order_date), strftime('%m', order_date)
)
SELECT order_month,
       SUM(CASE WHEN order_year = '2022' THEN sales ELSE 0 END) AS sales_2022,
       SUM(CASE WHEN order_year = '2023' THEN sales ELSE 0 END) AS sales_2023,
       SUM(CASE WHEN order_year = '2023' THEN sales ELSE 0 END) - 
       SUM(CASE WHEN order_year = '2022' THEN sales ELSE 0 END) AS difference
FROM cte
GROUP BY order_month
ORDER BY order_month;
'''
result = pd.read_sql(q4, conn)
print(result)
```
